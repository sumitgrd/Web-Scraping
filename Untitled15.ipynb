```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_wikipedia_headers(url):
    # Send a GET request to the Wikipedia page
    response = requests.get(url)
    
    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find all header tags (h1 to h6)
        headers = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
        
        # Extract the text content of the headers
        header_texts = [header.text.strip() for header in headers]
        
        return header_texts
    else:
        print(f"Error: Unable to fetch content from {url}")
        return None

def main():
    wikipedia_url = 'https://en.wikipedia.org/wiki/Main_Page'
    
    # Get the header tags from Wikipedia
    headers = get_wikipedia_headers(wikipedia_url)
    
    if headers:
        # Create a DataFrame from the header texts
        df = pd.DataFrame({'Header Tags': headers})
        
        # Display the DataFrame
        print(df)

if __name__ == "__main__":
    main()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Scrape top 10 ODI teams
teams_url = 'https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
teams_response = requests.get(teams_url)
teams_soup = BeautifulSoup(teams_response.content, 'html.parser')

team_rows = teams_soup.select('table.table tbody tr')
team_data = []
for row in team_rows:
  rank = row.select_one('.rank-cell').text.strip()
  team = row.select_one('.u-hide-phablet').text.strip()
  matches = row.select_one('.matches-cell').text.strip()
  points = row.select_one('.points-cell').text.strip()
  rating = row.select_one('.rating').text.strip()
  team_data.append([rank, team, matches, points, rating])

teams_df = pd.DataFrame(team_data, columns=['Rank', 'Team', 'Matches', 'Points', 'Rating'])

# Scrape top 10 ODI batsmen
batsmen_url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting'
batsmen_response = requests.get(batsmen_url)
batsmen_soup = BeautifulSoup(batsmen_response.content, 'html.parser')

batsman_rows = batsmen_soup.select('table.table tbody tr')
batsman_data = []
for row in batsman_rows:
  rank = row.select_one('.rank-cell').text.strip()
  player = row.select_one('.table-body__cell.name a').text.strip()
  team = row.select_one('.table-body__cell.u-hide-phablet a').text.strip()
  rating = row.select_one('.rating').text.strip()
  batsman_data.append([rank, player, team, rating])

batsmen_df = pd.DataFrame(batsman_data, columns=['Rank', 'Player', 'Team', 'Rating'])

# Scrape top 10 ODI bowlers
bowlers_url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling'
bowlers_response = requests.get(bowlers_url)
bowlers_soup = BeautifulSoup(bowlers_response.content, 'html.parser')

bowler_rows = bowlers_soup.select('table.table tbody tr')
bowler_data = []
for row in bowler_rows:
  rank = row.select_one('.rank-cell').text.strip()
  player = row.select_one('.table-body__cell.name a').text.strip()
  team = row.select_one('.table-body__cell.u-hide-phablet a').text.strip()
  rating = row.select_one('.rating').text.strip()
  bowler_data.append([rank, player, team, rating])

bowlers_df = pd.DataFrame(bowler_data, columns=['Rank', 'Player', 'Team', 'Rating'])

# Print the dataframes
print("Top 10 ODI Teams:")
print(teams_df.head(10))
print("\nTop 10 ODI Batsmen:")
print(batsmen_df.head(10))
print("\nTop 10 ODI Bowlers:")
print(bowlers_df.head(10))
```

4) Write a python program to scrape cricket rankings from https://www.icc-cricket.com/rankings/team-rankings/mens/odi You have to scrape and make data frame-
a) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.
b) Top 10 women’s ODI Batting players along with the records of their team and rating.
c) Top 10 women’s ODI all-rounder along with the records of their team and rating.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_odi_team_rankings(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    teams_data = []
    for team in soup.select('.rankings-block__banner, .table-body__row')[:10]:
        team_name = team.select_one('.u-hide-phablet').text.strip()
        matches = team.select_one('.table-body__cell.u-center-text').text.strip()
        points = team.select('.table-body__cell.u-center-text')[1].text.strip()
        rating = team.select_one('.table-body__cell.u-text-right.rating').text.strip()

        teams_data.append({
            'Team': team_name,
            'Matches': matches,
            'Points': points,
            'Rating': rating
        })

    return pd.DataFrame(teams_data)

def scrape_odi_batting_players(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    players_data = []
    for player in soup.select('.table-body__row')[:10]:
        player_name = player.select_one('.table-body__cell.name').text.strip()
        team = player.select_one('.table-body__cell.u-text-left a').text.strip()
        rating = player.select_one('.table-body__cell.rating').text.strip()

        players_data.append({
            'Player': player_name,
            'Team': team,
            'Rating': rating
        })

    return pd.DataFrame(players_data)

def scrape_odi_allrounders(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    allrounders_data = []
    for allrounder in soup.select('.table-body__row')[:10]:
        allrounder_name = allrounder.select_one('.table-body__cell.name').text.strip()
        team = allrounder.select_one('.table-body__cell.u-text-left a').text.strip()
        rating = allrounder.select_one('.table-body__cell.rating').text.strip()

        allrounders_data.append({
            'Player': allrounder_name,
            'Team': team,
            'Rating': rating
        })

    return pd.DataFrame(allrounders_data)

if __name__ == "__main__":
    odi_team_rankings_url = "https://www.icc-cricket.com/rankings/team-rankings/womens/odi"
    odi_batting_players_url = "https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting"
    odi_allrounders_url = "https://www.icc-cricket.com/rankings/womens/player-rankings/odi/all-rounder"

    df_teams = scrape_odi_team_rankings(odi_team_rankings_url)
    df_batting_players = scrape_odi_batting_players(odi_batting_players_url)
    df_allrounders = scrape_odi_allrounders(odi_allrounders_url)

    print("Top 10 ODI Women's Teams:")
    print(df_teams)

    print("\nTop 10 Women's ODI Batting Players:")
    print(df_batting_players)

    print("\nTop 10 Women's ODI All-rounders:")
    print(df_allrounders)
```

5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and
make data frame-
i) Headline
ii) Time
iii) News Link

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_cnbc_world_news(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    news_data = []
    for news_item in soup.select('.Card-title'):
        headline = news_item.text.strip()
        time = news_item.find_next(class_='Card-time').text.strip()
        news_link = news_item.find_parent('a')['href']

        news_data.append({
            'Headline': headline,
            'Time': time,
            'News Link': news_link
        })

    return pd.DataFrame(news_data)

if __name__ == "__main__":
    cnbc_world_news_url = "https://www.cnbc.com/world/?region=world"

    df_cnbc_world_news = scrape_cnbc_world_news(cnbc_world_news_url)

    print("CNBC World News:")
    print(df_cnbc_world_news)
```

6) Write a python program to scrape the details of most downloaded articles from AI in last 90
days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
Scrape below mentioned details and make data frame-
i) Paper Title
ii) Authors
iii) Published Date
iv) Paper URL

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_most_downloaded_articles(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    articles_data = []
    for article in soup.select('.article-details'):
        title = article.select_one('.article-title').text.strip()
        authors = ', '.join(author.text.strip() for author in article.select('.author-name'))
        published_date = article.select_one('.text-xs').text.strip()
        paper_url = article.find_previous('a')['href']

        articles_data.append({
            'Paper Title': title,
            'Authors': authors,
            'Published Date': published_date,
            'Paper URL': paper_url
        })

    return pd.DataFrame(articles_data)

if __name__ == "__main__":
    ai_most_downloaded_url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"

    df_most_downloaded_articles = scrape_most_downloaded_articles(ai_most_downloaded_url)

    print("Most Downloaded Articles in AI (Last 90 Days):")
    print(df_most_downloaded_articles)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Send a GET request to the website
url = "https://www.dineout.co.in"
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the elements containing the details you want to scrape
restaurant_names = soup.find_all('h2', class_='restnt-name ellipsis')
cuisines = soup.find_all('span', class_='double-line-ellipsis')
locations = soup.find_all('span', class_='double-line-ellipsis')
ratings = soup.find_all('span', class_='rating-value')
image_urls = soup.find_all('img', class_='img-responsive')

# Create empty lists to store the scraped data
restaurant_list = []
cuisine_list = []
location_list = []
rating_list = []
image_url_list = []

# Extract the data from the elements and append them to the respective lists
for name in restaurant_names:
  restaurant_list.append(name.text.strip())

for cuisine in cuisines:
  cuisine_list.append(cuisine.text.strip())

for location in locations:
  location_list.append(location.text.strip())

for rating in ratings:
  rating_list.append(rating.text.strip())

for image in image_urls:
  image_url_list.append(image['src'])

# Create a dictionary from the lists
data = {
  'Restaurant Name': restaurant_list,
  'Cuisine': cuisine_list,
  'Location': location_list,
  'Ratings': rating_list,
  'Image URL': image_url_list
}

# Create a dataframe from the dictionary
df = pd.DataFrame(data)

# Print the dataframe
print(df)
```
